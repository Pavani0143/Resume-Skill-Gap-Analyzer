```python
required_skills = [
    "python",
    "sql",
    "excel",
    "power bi",
    "statistics",
    "data cleaning"
]
```

```python
resume_skills = [
    "python",
    "sql",
    "excel",
    "pandas"
]
```

```python
present_skills = set(required_skills) & set(resume_skills)
missing_skills = set(required_skills) - set(resume_skills)

print("Present Skills:", present_skills)
print("Missing Skills:", missing_skills)
```

```python
match_percentage = (len(present_skills) / len(required_skills)) * 100
print(f"Skill Match Percentage: {match_percentage:.2f}%")
```

```python
import matplotlib.pyplot as plt

labels = ["Present Skills", "Missing Skills"]
values = [len(present_skills), len(missing_skills)]

plt.bar(labels, values)
plt.title("Resume Skill Gap Analysis")
plt.ylabel("Number of Skills")
plt.show()
```

```python
import pandas as pd

skill_status = []

for skill in required_skills:
    if skill in resume_skills:
        skill_status.append({"Skill": skill.title(), "Status": "Present"})
    else:
        skill_status.append({"Skill": skill.title(), "Status": "Missing"})

skill_df = pd.DataFrame(skill_status)
skill_df
```

```python
present_count = skill_df[skill_df["Status"] == "Present"].shape[0]
missing_count = skill_df[skill_df["Status"] == "Missing"].shape[0]

print("Total Required Skills:", len(required_skills))
print("Skills Present:", present_count)
print("Skills Missing:", missing_count)
print(f"Skill Match Percentage: {match_percentage:.2f}%")
```

```python
import matplotlib.pyplot as plt

status_counts = skill_df["Status"].value_counts()

plt.figure(figsize=(6,4))
plt.bar(status_counts.index, status_counts.values)
plt.title("Resume Skill Gap Analysis")
plt.xlabel("Skill Status")
plt.ylabel("Number of Skills")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(status_counts.index, status_counts.values)
plt.title("Resume Skill Gap Analysis")
plt.xlabel("Skill Status")
plt.ylabel("Number of Skills")
plt.tight_layout()
plt.savefig("skill_gap_analysis.png")
plt.show()
```

```python
print("Conclusion:")
print(f"The candidate matches {match_percentage:.2f}% of the required skills.")
print("Recommended skills to learn next:")
for skill in missing_skills:
    print("-", skill.title())
```

```python
core_skills = ["python", "sql", "excel"]
supporting_skills = ["power bi", "statistics", "data cleaning"]
```

```python
core_weight = 0.7
supporting_weight = 0.3

core_present = len(set(core_skills) & set(resume_skills))
supporting_present = len(set(supporting_skills) & set(resume_skills))

core_score = (core_present / len(core_skills)) * core_weight
supporting_score = (supporting_present / len(supporting_skills)) * supporting_weight

final_score = (core_score + supporting_score) * 100

final_score
```

```python
print(f"Final Skill Match Score: {final_score:.2f}%")

print("\nScoring Logic:")
print(" Core skills contribute 70% of the score")
print(" Supporting skills contribute 30% of the score")
```

```python
labels = ["Core Skills", "Supporting Skills"]
values = [core_present, supporting_present]

plt.figure(figsize=(6,4))
bars = plt.bar(labels, values)

plt.title("Skill Contribution Analysis (7030 Weighting)")
plt.ylabel("Skills Matched")

# Add numbers on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, int(height),
             ha='center', va='bottom')

plt.show()
```

```python
print("FINAL CONCLUSION")
print("-" * 40)

print(f"Overall Skill Match Score: {final_score:.2f}%")

print("\nKey Insights:")
print(f"- Core skills matched: {core_present}/{len(core_skills)}")
print(f"- Supporting skills matched: {supporting_present}/{len(supporting_skills)}")

if supporting_present == 0:
    print("\nObservation:")
    print("The candidate has strong core technical skills but lacks supporting tools.")
    print("This indicates readiness for entry-level roles with clear upskilling areas.")

print("\nRecommended Next Skills to Learn:")
for skill in missing_skills:
    print(f"- {skill.title()}")

print("\nConclusion:")
print("By focusing on the missing supporting skills, the candidate can significantly improve job readiness and interview success.")
```
